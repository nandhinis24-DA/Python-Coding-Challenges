### Day 45 â€“ Built-in Functions and User-Defined Functions
### Project Title: Simple Calculator App
### Scenario:
 A company wants to provide users with a simple calculator to perform basic mathematical operations.
Tasks:
Create functions for addition, subtraction, multiplication, and division.


Ask the user to input two numbers and the operation they want to perform.


Call the appropriate function and display the result.


Use built-in functions like abs(), round(), and pow() within your solution.
Handle division by zero gracefully and inform the user.

```python
# Simple Calculator App

# Functions
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    if b == 0:
        return "Error: Cannot divide by zero!"
    return a / b

# Input from user
a = float(input("Enter first number: "))
b = float(input("Enter second number: "))

print("Choose operation: +  -  *  /")
op = input("Enter operation: ")

# Perform calculation
if op == "+":
    result = add(a, b)
elif op == "-":
    result = subtract(a, b)
elif op == "*":
    result = multiply(a, b)
elif op == "/":
    result = divide(a, b)
else:
    result = "Invalid operation!"

# Display result
print("Result:", result)

# Using built-in functions
print("Absolute value of result:", abs(result) if isinstance(result, (int, float)) else result)
print("Rounded result:", round(result, 2) if isinstance(result, (int, float)) else result)
print("Result raised to power 2:", pow(result, 2) if isinstance(result, (int, float)) else result)
```
