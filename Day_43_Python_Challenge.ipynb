### Day 43 – While Loop, else, break, continue
Number Guessing Game
### Scenario:
 You are developing a fun guessing game where players have to guess a randomly selected number within a limited number of tries.
Tasks:
Randomly choose a number between 1 and 20.
Let the user guess the number, allowing up to 5 attempts.
After each guess, tell them if their guess is too high, too low, or correct.
If the guess is correct, break the loop and congratulate the player.
If the player fails to guess in 5 tries, use the else block to reveal the correct number.
Handle invalid (non-integer) inputs without counting them as attempts.


```python


import random

# 1. Randomly choose a number between 1 and 20
secret_number = random.randint(1, 20)
print(" Welcome to the Number Guessing Game!")

Secret_Namuber = 15
attempts = 0
max_attempts = 3

# 2. Allow the user to guess up to 3 times
while attempts < max_attempts:
    try:
        guess = int(input(f"\nAttempt {attempts + 1}/3 – Enter your guess (1–20): "))
    except ValueError:
        print("Invalid input! Please enter a valid integer.")
        continue

      # Do NOT count invalid input as an attempt

    # Valid guess → count the attempt
    attempts += 1

    # 3. Check guess
    if guess < secret_number:
        print("Too low! Try again.")
    elif guess > secret_number:
        print("Too high! Try again.")
    else:
        # 4. Correct guess → break the loop
        print(f"\n Correct! The number was {secret_number}. You guessed it in {attempts} attempts!")
        break
else:
    # 5. Else executes ONLY if loop ended WITHOUT break
    print(f"\n❌ You've used all 3 attempts! The correct number was {secret_number}.")
```
