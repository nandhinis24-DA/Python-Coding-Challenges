### Day 42 – Operators, Conditional Statements
### Project Title: BMI Calculator
### Scenario:
 A health organization wants a tool to help users calculate their Body Mass Index (BMI) and understand their health category based on it.
Tasks:
Ask the user to enter their weight (in kilograms) and height (in meters).
Calculate their BMI using the formula BMI = weight / (height × height).
Classify the BMI as underweight, normal, overweight, or obese using conditional statements.
Display the result clearly.
Ensure the weight and height entered are positive numbers.
Solve it when you have free time or on off days  !! don't have to be in a hurry !!

```python
# 1. Ask the user for weight and height
weight = float(input("Enter your weight in kg: "))
height = float(input("Enter your height in meters: "))

# 2. Ensure values are positive
if weight <= 0 or height <= 0:
    print("Weight and height must be positive numbers!")
else:
    # 3. Calculate BMI
    bmi = weight / (height * height)

    # 4. Classify BMI
    if bmi < 18.5:
        category = "Underweight"
    elif 18.5 <= bmi < 24.9:
        category = "Normal weight"
    elif 25 <= bmi < 29.9:
        category = "Overweight"
    else:
        category = "Obese"

    # 5. Display result
    print(f"\nYour BMI is: {bmi:.2f}")
    print("Health Category:", category)
```
