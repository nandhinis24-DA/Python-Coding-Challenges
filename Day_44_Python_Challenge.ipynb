### Day 44 – range, for loop, enumerate, list comprehension
### Project Title: Employee Hours Report
### Scenario:
 You are analyzing student exam scores to generate performance reports.
Tasks:
Create a list of 10 random exam scores ranging from 0 to 100 using list comprehension.


Display each score with its index using the enumerate function.


Find all scores greater than 75 and display them.


Calculate and display the average score.

```python

import random

# 1. Create a list of 10 random exam scores (0–100) using list comprehension
scores = [random.randint(0, 100) for _ in range(10)]
print("Exam Scores:", scores)

# 2. Display each score with its index using enumerate
print("\nScores with Index:")
for index, score in enumerate(scores):
    print(f"Index {index}: Score = {score}")

# 3. Find all scores greater than 75 and display them
high_scores = [score for score in scores if score > 75]
print("\nScores greater than 75:", high_scores)

# 4. Calculate and display the average score
average_score = sum(scores) / len(scores)
print(f"\nAverage Score: {average_score:.2f}")
```
